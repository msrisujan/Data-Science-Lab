## Question1

```python
class Stack:
    def __init__(self):
        self.stack = []
    def push(self, data):
        self.stack.append(data)
    def pop(self):
        if len(self.stack) < 1:
            return None
        else:
            return self.stack.pop()
    def peek(self):
        if len(self.stack) < 1:
            return None
        else:
            return self.stack[-1]
    def printStack(self):
        print(self.stack)
    def isEmpty(self):
        return self.stack == []
    def size(self):
        return len(self.stack)
    
stack = Stack()
while True:
    print("\n1. Push\n2. Pop\n3. Peek\n4. Print\n5. Exit")
    choice = int(input("Enter your choice: "))
    if choice == 1:
        data = int(input("Enter the data: "))
        stack.push(data)
    elif choice == 2:
        data = stack.pop()
        if data == None:
            print("Stack is empty")
        else:
            print("Popped data is", data)
    elif choice == 3:
        data = stack.peek()
        if data == None:
            print("Stack is empty")
        else:
            print("Top data is", data)
    elif choice == 4:
        stack.printStack()
    elif choice == 5:
        break
    else:
        print("Invalid choice")
```

## Question 2

```python
class Queue:
    def __init__(self):
        self.queue = []
    def enqueue(self, data):
        self.queue.append(data)
    def dequeue(self):
        if len(self.queue) < 1:
            return None
        else:
            return self.queue.pop(0)
    def peek(self):
        if len(self.queue) < 1:
            return None
        else:
            return self.queue[0]
    def printQueue(self):
        print(self.queue)
    def isEmpty(self):
        return self.queue == []
    def size(self):
        return len(self.queue)
    
queue = Queue()
while True:
    print("\n1. Enqueue\n2. Dequeue\n3. Peek\n4. Print\n5. Exit")
    choice = int(input("Enter your choice: "))
    if choice == 1:
        data = int(input("Enter the data: "))
        queue.enqueue(data)
    elif choice == 2:
        data = queue.dequeue()
        if data == None:
            print("Queue is empty")
        else:
            print("Dequeued data is", data)
    elif choice == 3:
        data = queue.peek()
        if data == None:
            print("Queue is empty")
        else:
            print("Front data is", data)
    elif choice == 4:
        queue.printQueue()
    elif choice == 5:
        break
    else:
        print("Invalid choice")
```

## Question 3

```python
def longestValidParentheses(s):
    stack = []
    stack.append(-1)
    result = 0
    for i in range(len(s)):
        if s[i] == '(':
            stack.append(i)
        else:
            stack.pop()
            if len(stack) != 0:
                result = max(result, i - stack[-1])
            else:
                stack.append(i)
    return result
s = input("Enter the string: ")
print(longestValidParentheses(s))
```

## Question 5

```python
def replaceVowels(s):
    vowels = ['a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U']
    for i in range(len(s)):
        if s[i] in vowels:
            s[i] = chr(ord(s[i]) + 1)
    return s

def replaceS(s):
    for i in range(len(s)):
        if s[i] == 's' or s[i] == 'S':
            s[i] = '$'
    return s

def makeFirstCharUpper(s):
    for i in range(len(s)):
        if i == 0:
            s[i] = s[i].upper()
        elif s[i] == '.':
            if i<len(s)-1:
                if s[i + 1] == ' ':
                    s[i + 2] = s[i + 2].upper()
                else:
                    s[i + 1] = s[i + 1].upper()
    return s

def countWords(s):
    count = 0
    for i in range(len(s)):
        if s[i] == ' ':
            count += 1
    return count + 1

def countSentences(s):
    count = 0
    for i in range(len(s)):
        if s[i] == '.':
            count += 1
    return count

def checkLength(s):
    return len(s) >= 50

s = input("Enter the string: ")
if not checkLength(s):
    print("Length of string is less than 50")
    exit()
print("Length of string is", len(s))
print("Total number of words is", countWords(s))
print("Total number of sentences is", countSentences(s))
s = list(s)
s = makeFirstCharUpper(s)
s = replaceVowels(s)
s = replaceS(s)
s = "".join(s)
print("Modified string is", s)
```
